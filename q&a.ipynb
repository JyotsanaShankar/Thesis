```python
import os
import json
import fitz  # PyMuPDF
import PyPDF2
from pdf2image import convert_from_path
from haystack.document_stores.faiss import FAISSDocumentStore
#from haystack.reader import TransformersReader
from haystack.pipelines import DocumentSearchPipeline
from haystack.nodes import PDFToTextConverter
import tabula
from PIL import Image
import io
import tempfile

# Load questions and answers from a JSON file
#with open("question_answers.json", "r") as file:
 #   data = json.load(file)

questions = data["questions"]
answers = data["answers"]


# Index PDF files into the document store
pdf_dir = "pdf_files"
converted_files = []
pdf_converter = PDFToTextConverter(remove_numeric_tables= False, valid_languages=["en"])
all_documents = []
# Iterate through PDF files in the directory
for pdf_file in os.listdir(pdf_dir):
    if pdf_file.endswith(".pdf"):
        # Extract text from the PDF file
        pdf_path = os.path.join(pdf_dir, pdf_file)
        doc = fitz.open(pdf_path)
        pdf_text = ""
        print('reading', pdf_path, 'having', len(doc), 'pages' )

        for page_num in range(len(doc)):
            page = doc[page_num]
            pdf_text += page.get_text("text")

        # Extract images using PyMuPDF (fitz)
        images = []
        for page_num in range(len(doc)):
            page = doc[page_num]
            xref = page.get_pixmap()
            img_bytes = xref.samples

            # Save the image bytes to a temporary file
            with tempfile.NamedTemporaryFile(delete=False, suffix=".png") as img_file:
                img_file.write(img_bytes)
                img_file_name = img_file.name
                images.append(img_file_name)

        # Extract tabular data using tabula-py
        try:
            tabular_data = tabula.read_pdf(pdf_path, pages="all")
        except Exception as e:
            tabular_data = []

        # Initialize an empty list to store the converted tabular data
        tabular_data_dicts = []

        # Iterate through the list of DataFrames and convert each one to a list of dictionaries
        for df in tabular_data:
            if not df.empty:
                tabular_data_dicts.append(df.to_dict(orient='records'))

        #converted = pdf_converter.convert(file_path=pdf_path, meta={"file_name": pdf_file})
        #converted_files.append(converted)
        # Create a document and add it to the document store
        document = {
            "content": pdf_text,
            "images": images,
            "tabular_data": tabular_data_dicts,
            "meta": {"file_name": pdf_file}
        }
        all_documents.append(document)

#document_store.write_documents(all_documents)
```

```python
from haystack.nodes import PreProcessor
from haystack.document_stores import InMemoryDocumentStore

#document_store = InMemoryDocumentStore(use_bm25=True)

preprocessor = PreProcessor(split_by="word",
                            split_length=250,
                            split_overlap=10,
                            clean_whitespace=True,
                            clean_header_footer=True,
                            clean_empty_lines=True,
                            split_respect_sentence_boundary=True,)

converted_docs = []
for files in all_documents:
  converted_docs = converted_docs + preprocessor.process(files)

# Initialize a document store (you can choose between FAISS or Elasticsearch)
document_store = FAISSDocumentStore(faiss_index_factory_str="Flat", return_embedding=True)
document_store.delete_documents()
document_store.write_documents(converted_docs)
```

```python
data = {
	"questions" : [
		"How much plastic has been reduced in transportation packaging materials in year 2020?"
	],
	"answers" : {
		"What is the total greenhouse gas emissions (in CO2e) generated by the company during the reporting period?": [
			"2,062 metric tons",
			"52,128 metric tons"
		]
	}
}
questions = data["questions"]
answers = data["answers"]
```

```python
from haystack.pipelines import DocumentSearchPipeline, ExtractiveQAPipeline
from haystack.nodes import DensePassageRetriever
#from haystack.rankers import DensePassageRanker



# Initialize Dense Passage Retriever (DPR)
retriever = DensePassageRetriever(document_store=document_store,
                                  #query_embedding_model="facebook/dpr-question_encoder-single-nq-base",
                                  #passage_embedding_model="facebook/dpr-ctx_encoder-single-nq-base",
                                  use_gpu=True)

document_store.update_embeddings(retriever)
```

```python
qa_pipeline = ExtractiveQAPipeline(reader=reader, retriever=retriever)
```

```python
from farm.modeling.tokenization import Tokenizer
from farm.modeling.prediction_head import QuestionAnsweringHead
from farm.infer import Inferencer

# Initialize FARM reader components
tokenizer = Tokenizer.load("bert-base-uncased")
question_answering_head = QuestionAnsweringHead.load("models/question_answering_model")

# Initialize the ExtractiveQAPipeline
inferencer = Inferencer(processor=question_answering_head, task_type="question_answering", gpu=False)
qa_pipeline = ExtractiveQAPipeline(reader=inferencer, retriever=document_store)
```

```python
# Perform question answering
answers = []
for question in questions:
    result = qa_pipeline.run(query=question, params={"Retriever": {"top_k": 10}})
    answers.append(result)

    # Extract answers
    #answers = result[0]["answers"]
    #for answer in answers:
    #    print("Answer:", answer["answer"])
```

```python
from haystack.pipelines import ExtractiveQAPipeline
from haystack.utils import print_answers

p_extractive_premade = ExtractiveQAPipeline(reader=reader, retriever=bm25_retriever)
res = p_extractive_premade.run(
    query=questions[0], params={"Retriever": {"top_k": 20}, "Reader": {"top_k": 15}}
)
#print_answers(res, details="minimum")
```

```python
print("Query:", res["query"])
answer_list = res["answers"]
for answer in answer_list:
  #print("Answer:", answer["answers"][0])
  answer_text = answer.answer
  score = answer.score
  print('\n')
  print("Answer:", answer_text)
  print("Score:", score)
  print("Filename:", answer.meta['file_name'])
```
